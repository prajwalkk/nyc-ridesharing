```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
df = pd.read_csv('data\yellow_tripdata_2019-01.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.to_pickle("data\yellow_tripdata_2019-01.pkl")
```

```python
del df['congestion_surcharge']
```

```python
df.head()
```

```python
df.dropna(inplace=True)
len(df)
```

```python
df.head()
```

```python
drop_indices = df[(df.tpep_pickup_datetime == df.tpep_dropoff_datetime )|(df.PULocationID == df.DOLocationID)|(df.passenger_count <= 0)| (df.passenger_count >= 3) | (df.PULocationID == 0)| (df.DOLocationID == 0)|(df.PULocationID == 264)| (df.DOLocationID == 264)|(df.PULocationID == 265)| (df.DOLocationID == 265)|(df.trip_distance < 0.1) ].index

df.drop(drop_indices, inplace=True)
```

```python
len(df)
```

```python
df.index = np.arange(1, len(df) + 1)
```

```python
df.info
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
```

```python
df["pickup_hour"] = df.tpep_pickup_datetime.dt.hour
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,8))
sns.countplot(x="pickup_hour", data=df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="pickup_hour", data=df[df['PULocationID']==138])
plt.show()
```
